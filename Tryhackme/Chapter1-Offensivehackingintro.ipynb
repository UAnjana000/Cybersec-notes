### üîì What is Offensive Security?

**Offensive Security** refers to the proactive side of cybersecurity‚Äîwhere professionals simulate attacks to find and fix vulnerabilities **before** real attackers do. It's the hacker mindset put to good use.

Think of it as ethical hacking or red teaming. It includes:

* **Reconnaissance** (finding information about a target)
* **Vulnerability scanning**
* **Exploitation** (testing weak points)
* **Post-exploitation** (what an attacker could do after gaining access)

The goal? **Strengthen security through simulated attacks**.

### üß∞ What is Gobuster?

**Gobuster** is a tool used in offensive security to brute-force:

* URLs (directories on a web server)
* DNS subdomains
* S3 buckets
* Virtual hosts

It's super fast and written in Go.

---

### üñ•Ô∏è Common Bash Commands with Gobuster

#### üîπ 1. Directory brute-forcing (most common)

```bash
gobuster dir -u http://target.com -w /usr/share/wordlists/dirb/common.txt
```

* `dir`: mode for directories
* `-u`: URL
* `-w`: wordlist

#### üîπ 2. DNS subdomain brute-forcing

```bash
gobuster dns -d target.com -w /usr/share/wordlists/dns/subdomains-top1million-5000.txt
```

* `dns`: mode for subdomains
* `-d`: domain

#### üîπ 3. Virtual host discovery

```bash
gobuster vhost -u http://target.com -w /usr/share/wordlists/vhosts.txt
```

#### üîπ 4. Adding file extensions (e.g., `.php`)

```bash
gobuster dir -u http://target.com -w /usr/share/wordlists/dirb/common.txt -x php,html,txt
```

#### üîπ 5. Use custom headers (e.g., for authorization)

```bash
gobuster dir -u http://target.com -w wordlist.txt -H "Authorization: Bearer <token>"
```
